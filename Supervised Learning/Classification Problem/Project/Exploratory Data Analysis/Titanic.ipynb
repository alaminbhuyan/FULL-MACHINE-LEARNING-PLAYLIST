# Data cleaning

```
# filing missing value help of Pclass
plt.figure(figsize=(10,9))
sns.boxplot(x="Pclass",y="Age",data=df,palette="hot")
plt.show()
```

```
# make a function
def impute_age(cols): # here cols contains ['Age','Pclass']
  age = cols[0]    # cols[0] means Age
  pclass = cols[1] # cols[1] means Pclass
  if pd.isnull(age):
    if pclass == 1:
      return 37
    elif pclass == 2:
      return 29
    else:
      return 24
  else:
    return age
```

```
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1)
```

```
# After filling the missing value
plt.figure(figsize=(10,8))
sns.heatmap(data=df.isnull(),cmap="viridis",yticklabels=False,cbar=False)
plt.show()
```

```
# In cabin class 687 null values so we drop this column
df.drop(labels=['Cabin'],axis=1,inplace=True)
```

```
df.head()
```

```
# After droping the Cabin column
plt.figure(figsize=(10,8))
sns.heatmap(data=df.isnull(),cmap="viridis",yticklabels=False,cbar=False)
plt.show()
```

# Converting Categorical Features
**We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.**

```
pd.get_dummies(data=df.Embarked,drop_first=True).head()
```

```
Sex = pd.get_dummies(data=df.Sex,drop_first=True)
Embarked = pd.get_dummies(data=df.Embarked,drop_first=True)
```

```
df.columns
```

```
df.drop(labels=['PassengerId','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```
df = pd.concat([df,Sex,Embarked],axis=1)
df.head()
```

```
features = df.drop(["Survived"],axis=1)
level = df.Survived
```

```
features
```

```
level
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```
X_train,x_test,y_train,y_test = train_test_split(features,level,test_size=0.28,random_state=1)
```

```
model = LogisticRegression(max_iter=300).fit(X_train,y_train)
```

```
model.score(x_test,y_test)
```

```
y_pred = model.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix
```

```
plt.figure(figsize=(10,8))
axes = sns.heatmap(data=pd.DataFrame(confusion_matrix(y_test,y_pred)),annot=True,fmt="d",cbar=False)
axes.set_xlabel("Predicted value",weight='bold')
axes.set_ylabel("Actual value",weight='bold')
class_level = ["not_survived","survived"]
tick_marks = np.arange(len(class_level))+0.5
axes.set_xticks(tick_marks)
axes.set_xticklabels(class_level,rotation=0)
axes.set_yticks(tick_marks)
axes.set_yticklabels(class_level,rotation=0)
plt.show()
```

```
pd.crosstab(y_test,y_pred,rownames=["Actual value"],colnames=["Predicted value"],margins=True)
```
